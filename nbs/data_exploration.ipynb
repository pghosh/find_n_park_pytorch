```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
```

```python
df=pd.read_csv("../data/Annual_Parking_Study_Data.csv")
df.shape
```

```python
df.dropna(inplace=True)
df.shape
```

```python
sns.set(style="darkgrid")
ax = sns.countplot(x="Study_Area", data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

```python
df_bt=df[df.Study_Area=='Belltown']
df_bt.head()
```

```python
plt.figure(figsize=(20,10),dpi=100)
sns.set(style="darkgrid")
ax = sns.countplot(x="Unitdesc", data=df_bt)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```
